## Matplotlib

<b>Matplotlib</b> es la librería de Python para realizar gráficas de funciones matemáticas.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd  
```

```python
x = np.arange(0,10,0.1) #Puntos a evaluar almacenados en el vector x
y = x*np.cos(x)         #Imagen para cada uno de los puntos en el vector x de la función  
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('f(x)')
plt.show()
```
